### Option 1: Use Pandas to create the contacts DataFrame.

```python
import pandas as pd
import json


# Extract the contact information from the temporary column
dict_values = []
for i, row in contact_info_df.iterrows():
    data = row['contact_info']
    try:
        converted_data = json.loads(data)
        dict_values.append([converted_data.get('contact_id', ''), converted_data.get('name', ''), converted_data.get('email', '')])
    except json.JSONDecodeError:
        # Handle the case where JSON decoding fails
        dict_values.append(['', '', ''])

# Create a new DataFrame with the extracted values
columns = ['contact_id', 'first_name', 'last_name', 'email']
extracted_df = pd.DataFrame(dict_values, columns=columns)

# Display the extracted DataFrame
print(extracted_df)
```

```python
# Check the datatypes.
data_types = contact_info_df.dtypes

# Print the data types
print(data_types)
```
